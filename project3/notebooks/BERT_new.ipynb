```python
import pandas as pd
import transformers
import torch

# Load pre-trained BERT model and tokenizer
model = transformers.BertModel.from_pretrained('bert-base-uncased')
tokenizer = transformers.BertTokenizer.from_pretrained('bert-base-uncased')

```

```python
data= pd.read_csv("C:\\Users\\agnes\\Documents\\apziva_ai_residency\\project3\\data\\potential-talents.csv")
words=data.job_title
print(words)
```

```python

def string_to_bert_embedding(string):
    # Tokenize and encode the string
    input_ids = tokenizer.encode_plus(string, add_special_tokens=True, return_tensors='pt')
    # Get the output of the BERT model
    output = model(input_ids)
    # Compute the average of the last hidden states
    embedding = torch.mean(output.last_hidden_state, dim=1)
    #print(embedding)
    embedding = embedding.detach().numpy()
    return embedding
#new_list=[""]*len(data["job_title"])
data["embedding"]=data["job_title"].apply(lambda x: string_to_bert_embedding(str(x)))
```
